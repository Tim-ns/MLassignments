```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
# load raw data
x_test = pd.read_csv("X_test.csv")
x_train = pd.read_csv("X_train.csv")

# create dataframe
x_test = pd.DataFrame(x_test)
x_train = pd.DataFrame(x_train)

# data size
x_test.shape, x_train.shape
```

```python
# check if the data is not null
x_test_notnull = x_test.notnull()
x_train_notnull = x_train.notnull()

x_test_notnull, x_train_notnull.shape
```

```python
# clean the zero data after the collision
# Since we know that after the collision, the position, velocity, and time for that trajectory will become zero.

# Dropping rows where all columns are equal to 0
x_train_cleaned = x_train[(x_train.drop(columns=['Id']) != 0).any(axis=1)]

# Show the cleaned DataFrame
#x_train_cleaned
x_train_cleaned.shape
```

```python
# Histogram of x_train and x_train_cleaned

# Histogram of x_train.x_1
plt.figure(figsize=(5, 3))

plt.hist(x_train.x_1, bins=300, edgecolor='k', alpha=0.7)
plt.title('Histogram of x_train.x_1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Histogram of x_train_cleaned.x_1
plt.hist(x_train_cleaned.x_1, bins=300, edgecolor='k', alpha=0.7)
plt.title('Histogram of x_train_cleaned.x_1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Histogram of x_train.x_2
plt.hist(x_train.x_2, bins=300, edgecolor='k', alpha=0.7)
plt.title('Histogram of x_train.x_2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Histogram of x_train_cleaned.x_2
plt.hist(x_train_cleaned.x_2, bins=300, edgecolor='k', alpha=0.7)
plt.title('Histogram of x_train_cleaned.x_2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
